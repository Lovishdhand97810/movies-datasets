```python
surplus = a[a['Amount_y'] > a['Amount_x']]
surplus
```

```python
surplus['difference'] = a['Amount_y'] - a['Amount_x']
surplus
```

```python
surplus['return'] = surplus['difference'] + (surplus['difference']*(2/100))
#surplus['return'] = surplus['difference'].apply(Lambda x:(x*2)/100+x)
surplus
```

```python
surplus.rename(columns={'Amount_x': 'spent', 'Amount_y': 'repay'}, inplace=True)
```

```python
spend['date'] = pd.to_datetime(spend['Month'])
spend['date'] = spend['date'].dt.to_period('M')
```

```python
monthly_spend = pd.DataFrame(spend.groupby(['Customer','date'])['Amount'].sum()).reset_index(drop=False)
```

```python
monthly_spend
```

```python
max_spending = monthly_spend.sort_values(by = 'Amount',ascending=False).head(10)
```

```python
max_spending
```

```python
seg_max = pd.merge(max_spending,credit_banking[['Customer','Segment']], on = 'Customer',how = 'inner')
seg_max.drop('Customer',inplace = True,axis=1)
```

```python
seg_max
```

```python
seg_max.sort_values(by = 'Amount',ascending=False).head(10)
```

```python
seg_max_final = pd.DataFrame(seg_max.groupby('Segment')['Amount'].sum()).reset_index(drop = False)
seg_max_final.sort_values('Amount',ascending = False)
seg_max_final
```

```python
import matplotlib.pyplot as plt
```

```python
# Plotting a bar chart
plt.figure(figsize=(10, 6))
plt.bar(seg_max_final['Segment'], seg_max_final['Amount'], color='green', alpha=0.9)

# Adding labels and title
plt.xlabel('Segment', fontsize=12)
plt.ylabel('Total Amount', fontsize=12)
plt.title('Total Amount by Segment', fontsize=14)
plt.xticks(rotation=0, ha='right')  # Rotate segment names for better readability

# Displaying the plot
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(8,8))
plt.pie(
    seg_max_final['Amount'],
    labels = seg_max_final['Segment'],
    autopct = '%1.1f%%',
    startangle = 90,
    colors = plt.cm.tab10.colors
)

plt.title('Distribution of Total Amount by Segment',fontsize=14)
plt.show()
```
